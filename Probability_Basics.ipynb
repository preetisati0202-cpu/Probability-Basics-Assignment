Question 4: The population of a school is divided into 60% boys and 40% girls. If you want equal representation
of both genders in the sample, which method should you use: Simple Random Sampling or Stratified Sampling?
Why?

Answer 4: Population:

Boys = 60%
Girls = 40%

Correct Method: Stratified Sampling

WHY?

-The population has different groups (boys & girls)

-We want equal representation from both groups

-Stratified sampling divides population into groups (strata) and selects from each

Explanation

In simple random sampling, selection is random → may not give equal boys and girls

In stratified sampling, we:

-divide into boys and girls

-take equal sample from both

Stratified sampling, because it ensures equal representation of both boys and girls.



Question 5: The average height of 1000 students = 160 cm. A sample of 100 students shows an average height
= 158 cm.

Find the sampling error.

Answer 5: Population Mean = 160 cm

Sample mean = 150 cm

Sampling Error = Sample mean - Population

= 158-160

= -2 cm .

Question 10: Differentiate between discrete and continuous random variables with examples.

Answer 10: Discrete Random Variable

*   Takes countable values (0,1,2,3...)
*   Values are separate (no in between)   
*   Usually counting

Examples:


*   Number od students in a class
*   Number of heads in coin toss
*   Number of cars passing

Continous Random Variable

*   Takes any value in a range (decimals allowed)
*   Values are continuous (in-between possible)
*   Usually measuring

Examples:

*   Height of a person (e.g., 160.5 cm)
*   Weight of a person
*   Temperature.
